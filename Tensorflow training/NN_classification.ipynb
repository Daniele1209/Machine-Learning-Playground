### Create the data

```python
# Imports
from sklearn.datasets import make_circles
import pandas as pd
import seaborn as sns
```

```python
# create circle samples
circle_amount = 1000

X, y = make_circles(circle_amount, noise=0.03, random_state=42)
```

```python
X[:5]
```

```python
y[:5]
```

```python
circles = pd.DataFrame({'X0': X[:, 0], 'X1': X[:, 1], 'label': y})
circles.head(5)
```

```python
sns.scatterplot(
    data=circles,
    x=X[:, 0],
    y=X[:, 1],
    hue='label'
)
```
